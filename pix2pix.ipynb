#画像を任意のサイズの正方形にリサイズ

```python
import os
import cv2

def convert_to_square_images(input_dir, output_dir):
    # 出力ディレクトリが存在しない場合は作成する
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 入力ディレクトリ内の画像ファイルを取得
    image_files = [f for f in os.listdir(input_dir) if os.path.isfile(os.path.join(input_dir, f))]

    for i, filename in enumerate(image_files):
        input_path = os.path.join(input_dir, filename)
        img = cv2.imread(input_path)

        # 画像を任意の大きさにリサイズ
        resized_img = cv2.resize(img, (256, 256))

        # 保存するファイルパスを構築
        output_filename = str(i+1) + ".jpg"
        output_path = os.path.join(output_dir, output_filename)

        # リサイズした画像を保存
        cv2.imwrite(output_path, resized_img)
        print(f"Saved {output_filename}")

    print("Image conversion completed.")

# 入力ディレクトリと出力ディレクトリを指定
input_directory = "/content/make_BWtoCOLOR_datasets/archive"
output_directory = "/content/make_BWtoCOLOR_datasets/archive_out"

# 正方形に変換して画像を保存
convert_to_square_images(input_directory, output_directory)
```

#画像をグレースケール化して別のディレクトリに保存

```python
import os
import cv2

def convert_to_grayscale(input_dir, output_dir):
    # 出力ディレクトリが存在しない場合は作成する
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # 入力ディレクトリ内のファイルを順番に処理する
    for filename in os.listdir(input_dir):
        if filename.endswith(".jpg"):
            # 画像のパスを取得
            image_path = os.path.join(input_dir, filename)

            # カラー画像を読み込む
            color_image = cv2.imread(image_path)

            # グレースケール化
            gray_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)

            # 出力ファイルパスを生成
            output_path = os.path.join(output_dir, filename)

            # グレースケール画像を保存
            cv2.imwrite(output_path, gray_image)

            print(f"Converted {filename} to grayscale and saved as {output_path}")

# ディレクトリAとディレクトリBのパスを指定して変換を実行
input_directory = "/content/make_BWtoCOLOR_datasets/archive_out"
output_directory = "/content/make_BWtoCOLOR_datasets/images_gray"

convert_to_grayscale(input_directory, output_directory)
```

```python
import os
import random
import shutil

def split_images(img_dir, train_dir, test_dir, val_dir, train_ratio, test_ratio, val_ratio):
    # 古いディレクトリが存在する場合は削除する
    if os.path.exists(train_dir):
        shutil.rmtree(train_dir)
    if os.path.exists(test_dir):
        shutil.rmtree(test_dir)
    if os.path.exists(val_dir):
        shutil.rmtree(val_dir)

    # ディレクトリを作成
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # 画像ファイルのリストを取得し、名前順にソート
    image_files = sorted([f for f in os.listdir(img_dir) if f.endswith(".jpg")], key=lambda x: int(os.path.splitext(x)[0]))

    # 画像の総数と各セットの数を計算
    num_images = len(image_files)
    num_train = int(num_images * train_ratio)
    num_test = int(num_images * test_ratio)
    num_val = num_images - num_train - num_test

    # 画像を分割してコピー
    for i, filename in enumerate(image_files):
        src_path = os.path.join(img_dir, filename)

        if i < num_train:
            dst_dir = train_dir
        elif i < num_train + num_test:
            dst_dir = test_dir
        else:
            dst_dir = val_dir

        dst_path = os.path.join(dst_dir, filename)
        shutil.copy(src_path, dst_path)

        print(f"{filename} を {dst_dir} に移動しました")

    # 各ディレクトリに分けられた画像の枚数を表示
    print("各ディレクトリの画像枚数:")
    print(f"Train: {num_train}")
    print(f"Test: {num_test}")
    print(f"Validation: {num_val}")

# パラメータを設定
img_dir = "/content/make_BWtoCOLOR_datasets/archive_out"
train_dir = "/content/pytorch-CycleGAN-and-pix2pix/datasets/A/train/"
test_dir = "/content/pytorch-CycleGAN-and-pix2pix/datasets/A/test/"
val_dir = "/content/pytorch-CycleGAN-and-pix2pix/datasets/A/val/"
train_ratio = 0.8
test_ratio = 0.1
val_ratio = 0.1

# 画像を分割して配置し、枚数を表示
split_images(img_dir, train_dir, test_dir, val_dir, train_ratio, test_ratio, val_ratio)
```

```python
import os

def count_files_in_directory(directory_path):
    try:
        # 指定されたディレクトリ内のファイルリストを取得
        files_list = os.listdir(directory_path)

        # ファイルの数をカウント
        file_count = len(files_list)
        return file_count

    except OSError as e:
        print(f"エラー: {e}")
        return -1

# ディレクトリのパスを指定してファイル数を数える例
directory_path = "/content/pytorch-CycleGAN-and-pix2pix/datasets/AB/train"
file_count = count_files_in_directory(directory_path)

if file_count >= 0:
    print(f"ディレクトリ'{directory_path}'内のファイル数: {file_count}")
else:
    print("ファイル数を数える際にエラーが発生しました。")
```

# テスト画像に対する定性評価（SSIM）

```python
# テストの生成画像と正解画像が格納されたディレクトリへのパス
image_dir = './results/my_pix2pix_0727_2/test_latest/images'

# 画像ファイルのリストを取得
image_files = os.listdir(image_dir)

# 生成画像と正解画像を分けて保持する
real_images = [f for f in image_files if '_real_B.png' in f]
fake_images = [f for f in image_files if '_fake_B.png' in f]

# リスト内を整列する
real_images.sort()
fake_images.sort()

ssim_values = []

# 各画像ペアについてSSIMの値を計算する
for real_image_file, fake_image_file in zip(real_images, fake_images):
    real_image = cv2.imread(os.path.join(image_dir, real_image_file))
    fake_image = cv2.imread(os.path.join(image_dir, fake_image_file))

    # 各チャンネルについてSSIMを求める
    ssim_per_channel = cv2.quality.QualitySSIM_compute(real_image, fake_image)

    # チャンネルごとのSSIMの平均値を求める
    ssim = np.mean(ssim_per_channel[0]) #ssim_per_channel[0]にはRGB各チャンネルのSSIMがリストで格納されている、
                                        # ここではそのリストの平均を計算してssimに格納している
    ssim_values.append(ssim)

    # 各画像のSSIMを表示
    print(f'SSIM for {real_image_file} vs {fake_image_file}: {ssim}')

# SSIMの平均値を求める
average_ssim = np.mean(ssim_values)
print(f'Average SSIM: {average_ssim}')

# ヒストグラムの作成
plt.hist(ssim_values, bins=30, range=(0, 1))  # binsは階級の数、rangeはx軸の範囲
plt.xlabel('SSIM')
plt.ylabel('Frequency')
plt.title('SSIM Histogram')
plt.xticks([i * 0.1 for i in range(11)])  # 0から1までを0.1刻みでメモリを表示
plt.show()
```
